```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('IMDB_Movies.csv')
print(df.shape)
df
```

```python
df.columns
```

```python
df.drop(['color','num_critic_for_reviews','director_facebook_likes','actor_3_facebook_likes','actor_1_facebook_likes','cast_total_facebook_likes','facenumber_in_poster','plot_keywords','movie_imdb_link','content_rating','actor_2_facebook_likes','num_user_for_reviews','aspect_ratio','actor_2_name','actor_3_name'], axis = 1, inplace = True)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
director_name = df[(df['director_name'].isna() == True)]
df.drop(director_name.index, inplace = True)
```

```python

for i in df[(df['language'].isna() == True)]['director_name']:
    if len(df[df['director_name'] == i]['language'].value_counts()) != 0:
        language = df[(df['director_name'] == i)]['language'].value_counts().index[0]
        df.loc[(df['language'].isna() == True)&(df['director_name'] == i),'language'] = language
    else:
        print('empty')
```

```python
df.loc[df['language'].isna() == True,'language'] = 'English'
```

```python
null = df.loc[(df['duration'].isna() == True)&(df['gross'].isna() == True)&(df['budget'].isna() == True)]
print(null.shape)
df.drop(null.index, inplace = True)
```

```python
for i,j in zip(df[(df['duration'].isna() == True)]['country'],df[(df['duration'].isna() == True)]['language']):
    if i != 'Pakistan':
        mean = df[(df['country']==i)&(df['language']==j)]['duration'].mean()
        df.loc[(df['duration'].isna() == True)&(df['country']==i)&(df['language']==j),'duration'] = mean
    else:
        mean = df['duration'].mean()
        df.loc[(df['duration'].isna() == True)&(df['country']==i)&(df['language']==j),'duration'] = mean
```

```python
num = 0
for a,b in zip(df[(df['gross'].isna() == True)]['country'],df[(df['gross'].isna() == True)]['language']):
    if len(df.loc[(df['country']==a)&(df['language']==b)]['gross'].value_counts()) != 0:
        mean = df.loc[(df['country']==a)&(df['language']==b)]['gross'].mean()
        df.loc[(df['gross'].isna() == True)&(df['country']==a)&(df['language']==b),'gross'] = mean
    else:
        print('empty')
    num+=1
print(num)
    
```

```python
mean = df['gross'].mean()
df.loc[(df['gross'].isna() == True),'gross'] = mean
```

```python
num = 0
for a,b,c in zip(df[(df['budget'].isna() == True)]['country'],df[(df['budget'].isna() == True)]['language'],df[(df['budget'].isna() == True)]['director_name']):
    if len(df.loc[(df['country']==a)&(df['language']==b)&(df['director_name']==c)]['budget'].value_counts()) != 0:
        mean = df.loc[(df['country']==a)&(df['language']==b)&(df['director_name']==c)]['budget'].mean()
        df.loc[(df['budget'].isna() == True)&(df['country']==a)&(df['language']==b)&(df['director_name']==c),'budget'] = mean
    else:
        print('empty')
    num+=1
print(num)
```

```python
num = 0
for a,b in zip(df[(df['budget'].isna() == True)]['country'],df[(df['budget'].isna() == True)]['language']):
    if len(df.loc[(df['country']==a)&(df['language']==b)]['budget'].value_counts()) != 0:
        mean = df.loc[(df['country']==a)&(df['language']==b)]['budget'].mean()
        df.loc[(df['budget'].isna() == True)&(df['country']==a)&(df['language']==b),'budget'] = mean
    else:
        print('empty')
    num+=1
print(num)
```

```python
mean = df['budget'].mean()
df.loc[(df['budget'].isna() == True),'budget'] = mean
```

```python
df.loc[(df['country'].isna() == True),'country'] = 'USA'
```

```python
null = df.loc[df['actor_1_name'].isna()==True]
df.drop(null.index, inplace = True)
```

```python
df['title_year'] = df['title_year'].fillna(df['title_year'].median())
```
